```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
import joblib

path = r'C:\Users\user\Desktop\Revature\Projects\Yelp\yelp_reviews_sample.csv'

df = pd.read_csv(path)
```

```python
stop_words = set(stopwords.words('english')) #'if', 'and', 'the', etc.

def preprocess(text):
    translation = str.maketrans('', '', string.punctuation)
    text = text.translate(translation)
    text = text.lower()
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

df['text'] = df.apply(lambda row: preprocess(row['text']), axis=1)

print(df.head())
```

```python
vectorizer = TfidfVectorizer()

vectors = vectorizer.fit_transform(df.text)
joblib.dump(vectors, "vectors.joblib")
```

```python
NB_model = MultinomialNB()
NB_model = NB_model.fit(vectors, df.stars)

df_test = vectorizer.transform(df.text)
predictions = NB_model.predict(df_test)
print(accuracy_score(df.stars, predictions))
```

```python
joblib.dump(NB_model, "NB_model.joblib")
```

```python
SGD_class = SGDClassifier()
SGD_class = SGD_class.fit(vectors, df.stars)

predictions = SGD_class.predict(df_test)
print(accuracy_score(df.stars, predictions))
```

```python
joblib.dump(SGD_class, "SGD_class.joblib")
```

```python
linearSVC = LinearSVC()
linearSVC = linearSVC.fit(vectors, df.stars)

predictions = linearSVC.predict(df_test)
print(accuracy_score(df.stars, predictions))
```

```python
joblib.dump(linearSVC, "linearSVC.joblib")
```
